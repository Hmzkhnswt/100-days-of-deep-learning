```python
import pandas as pd
import numpy as np
import seaborn as sns
```

### Importing Sample Data

```python
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=1000,noise=0.1, random_state=41)
```

```python
X.shape
```

```python
sns.scatterplot(x = X[:,0], y = X[:,1], hue=y)
```

```python
import tensorflow as tf
import keras
from keras.layers import Dense,BatchNormalization
from keras.models import Sequential
```

### Using Batch NOrmalization()

```python
model = Sequential()
model.add(Dense(12, activation='relu', input_dim = 2))
model.add(BatchNormalization())
model.add(Dense(9, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(6, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(3, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X,y, epochs=200,validation_split=0.2)
```

### Without BAtch Normalization

```python
model1 = Sequential()
model1.add(Dense(12, activation='relu', input_dim = 2))

model1.add(Dense(9, activation='relu'))

model1.add(Dense(6, activation='relu'))

model1.add(Dense(3, activation='relu'))

model1.add(Dense(1, activation='sigmoid'))

model1.summary()
```

```python
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history1 = model1.fit(X,y, epochs=200,validation_split=0.2)
```

### Comparison Between the BAtch Normalization and Without BAtch Normalization

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'],label = 'BatchNormalization')
plt.plot(history1.history['accuracy'], label = 'No BatchNormalizatio')
plt.xlabel('Epochs')
plt.ylabel('Training Accuracy')
plt.title('Training Accuracy Comparison')
plt.legend()
```
